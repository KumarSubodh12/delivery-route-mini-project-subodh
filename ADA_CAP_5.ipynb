## Delivery Route Optimization using Algorithmic Strategies

## Input Modeling

### Explanation
We define locations, distances, parcels, and vehicle constraints.

```python
# Distance Matrix (Graph)
graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

# Parcels (value, weight)
parcels = [(60, 10), (100, 20), (120, 30)]

# Vehicle capacity
capacity = 50
```

## Recurrence - Route Cost

```python
def route_cost(graph, path):
    cost = 0
    for i in range(len(path)-1):
        cost += graph[path[i]][path[i+1]]
    cost += graph[path[-1]][path[0]]
    return cost

print("Route Cost:", route_cost(graph, [0,1,2,3]))
```

## Greedy - Parcel Selection

```python
def greedy_knapsack(parcels, capacity):
    parcels.sort(key=lambda x: x[0]/x[1], reverse=True)

    total_value = 0
    for value, weight in parcels:
        if capacity >= weight:
            capacity -= weight
            total_value += value

    return total_value

print("Max Value:", greedy_knapsack(parcels, capacity))
```

## Dynamic Programming - Delivery Planning

```python
def dp_knapsack(W, wt, val, n):
    dp = [[0]*(W+1) for _ in range(n+1)]

    for i in range(1, n+1):
        for w in range(W+1):
            if wt[i-1] <= w:
                dp[i][w] = max(val[i-1] + dp[i-1][w-wt[i-1]], dp[i-1][w])
            else:
                dp[i][w] = dp[i-1][w]

    return dp[n][W]

wt = [10,20,30]
val = [60,100,120]

print("DP Result:", dp_knapsack(capacity, wt, val, len(wt)))
```

## Shortest Path - Dijkstra

```python
import heapq

def dijkstra(graph, start):
    n = len(graph)
    dist = [float('inf')] * n
    dist[start] = 0
    pq = [(0, start)]

    while pq:
        d, u = heapq.heappop(pq)

        for v in range(n):
            if graph[u][v] > 0:
                if dist[v] > d + graph[u][v]:
                    dist[v] = d + graph[u][v]
                    heapq.heappush(pq, (dist[v], v))

    return dist

print("Shortest Paths:", dijkstra(graph, 0))
```

```python
def prim(graph):
    n = len(graph)
    visited = [False]*n
    visited[0] = True
    cost = 0

    for _ in range(n-1):
        min_edge = float('inf')
        x = y = 0

        for i in range(n):
            if visited[i]:
                for j in range(n):
                    if not visited[j] and graph[i][j]:
                        if graph[i][j] < min_edge:
                            min_edge = graph[i][j]
                            x, y = i, j

        cost += min_edge
        visited[y] = True

    return cost

print("MST Cost:", prim(graph))
```

## Travelling Salesman Problem

```python
from itertools import permutations

def tsp(graph):
    n = len(graph)
    nodes = list(range(n))
    min_cost = float('inf')

    for perm in permutations(nodes[1:]):
        cost = 0
        k = 0

        for j in perm:
            cost += graph[k][j]
            k = j

        cost += graph[k][0]
        min_cost = min(min_cost, cost)

    return min_cost

print("TSP Cost:", tsp(graph))
```

```python
import time

start = time.time()
tsp(graph)
end = time.time()

print("TSP Time:", end - start)
```
